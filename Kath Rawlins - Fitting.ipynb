```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import data from starbucks_drinkMenu_expanded.csv
import pandas as pd
df = pd.read_csv('starbucks_drinkMenu_expanded.csv')
df.head()
```

```python
# print the keys of the dataframe
df.keys()
```

```python
# print Beverage
# df[' Calcium (% DV) ']

xdata = df[' Calcium (% DV) ']
ydata = df[' Protein (g) ']

# remove units and other characters from the data
xdata = xdata.str.replace('%', '')
xdata = xdata.str.replace(' ', '')
print(type(ydata[1]))
x = np.zeros(len(xdata))
y = np.zeros(len(ydata))
# put x and y into numpy arrays
for i in range(len(xdata)):
    x[i] = float(xdata[i])
    y[i] = float(ydata[i])
print(x[10])
print(y[10])
```

```python
plt.figure(figsize=(24, 12), layout='constrained')
plt.scatter(x,y)

plt.legend(["legend"], fontsize='smaller', loc='upper right', bbox_to_anchor=(1.1, 1.1))
# scale axes to be logarithmic
# plt.xscale('log')
plt.yscale('log')

# x = xdata
# y = ydata
#fit polynomial models up to degree 5
# model1 = np.poly1d(np.polyfit(x, y, 1))
model2 = np.poly1d(np.polyfit(x, y, 2))
model3 = np.poly1d(np.polyfit(x, y, 3))
model4 = np.poly1d(np.polyfit(x, y, 4))
model5 = np.poly1d(np.polyfit(x, y, 5))

#create scatterplot
polyline = np.linspace(.95 * min(x), 1.05 * max(x), 50)
# plt.scatter(df.x, df.y)

#add fitted polynomial lines to scatterplot 
# plt.plot(polyline, model1(polyline), color='green')
plt.plot(polyline, model2(polyline), color='red')
plt.plot(polyline, model3(polyline), color='purple')
plt.plot(polyline, model4(polyline), color='blue')
plt.plot(polyline, model5(polyline), color='orange')

# plt.yscale('linear')
plt.show()
```

```python
# print the quality of the fit
from sklearn.metrics import r2_score
# print(r2_score(y, model1(x)))
print(r2_score(y, model2(x)))
```
